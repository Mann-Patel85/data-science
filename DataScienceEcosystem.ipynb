```python
#This will convert 200 minutes to hours by diving by 60
# The total minutes to convert
total_minutes = 200

# Convert total minutes to hours by dividing by 60.0 to ensure a float result
hours = total_minutes / 60.0

# Print the result
print(f"{total_minutes} minutes is equal to {hours} hours")

# Optional: To get an integer hour and remaining minutes
integer_hours = total_minutes // 60
remaining_minutes = total_minutes % 60
print(f"Alternatively, this is {integer_hours} hours and {remaining_minutes} minutes")
```

**Objectives:**
* Popular Languages for Data Science:
* Python: Dominant due to its simplicity, readability, and a vast ecosystem of libraries (e.g., Pandas, NumPy, Scikit-learn, TensorFlow, PyTorch)
* R: A language and environment specifically designed for statistical computing and graphics, favored in academia and research for its powerful visualization capabilities (e.g., ggplot2) and extensive statistical packages.
* SQL: Essential for managing and querying structured data in relational databases, a critical first step in the data science process for data retrieval and initial filtering.
* Julia: A high-performance language gaining traction for numerical and scientific computing due to its speed, often seen as a potential successor to Python or R for computationally intensive tasks.
* Scala: Frequently used in big data processing, particularly with frameworks like Apache Spark, due to its ability to run on the Java Virtual Machine (JVM) and provide speed and scalability.
