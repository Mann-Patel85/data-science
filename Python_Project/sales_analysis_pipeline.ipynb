```python
import numpy as np
import pandas as pd
import csv
```

```python
raw_data = """
ID|Product|Category|Sales|Region
101|Laptop|Electronics|$1000|North
102|Chair |Furniture|  200|north
103|Phone|electronics|$500|South
104|Table|Furniture|?|East
105|Hdd|Electronics|80|West
106|Headphones|  Electronics  |150|West
"""
```

```python
lines=raw_data.strip().split('\n')
clean_rows=[]
header=lines[0].split('|')
clean_rows=[header]
```

```python
for line in lines[1:]:
    parts=line.split('|')
    
    p_id=parts[0].strip()
    product=parts[1].strip()
    category=parts[2].strip().title()
    sales=parts[3].strip()
    if sales=='?':
        continue
    sales=int(sales.replace('$',''))
    region=parts[4].strip().title()
    
    clean_rows.append([p_id, product, category, sales, region])
print(clean_rows)
```

```python
with open('cleaned_sales.csv','w',newline='')as f:
    writer=csv.writer(f)
    writer.writerows(clean_rows)
print('File Saved')
```

```python
import pandas as pd
import numpy as np

# 1. Load the clean file
df = pd.read_csv('cleaned_sales.csv')

# 2. NumPy Logic (The "Advanced" Part)
# If Category is 'Electronics', Tax is 18% (0.18), else it is 12% (0.12)
df['Tax_Rate'] = np.where(df['Category'] == 'Electronics', 0.18, 0.12)

# Calculate the tax amount
df['Tax_Amount'] = df['Sales'] * df['Tax_Rate']

# 3. Pandas Analysis
print("--- Total Sales by Region ---")
print(df.groupby('Region')['Sales'].sum())

print("\n--- Final Data ---")
print(df)
```
